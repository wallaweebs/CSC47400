```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fah_sheet1 = pd.read_excel('Appendix B.xls', sheet_name='94-98 FAH')
fah_sheet2 = pd.read_excel('Appendix B.xls', sheet_name='03-04 FAH')
fah_sheet3 = pd.read_excel('Appendix B.xls', sheet_name='05-06 FAH')
fah_sheet4 = pd.read_excel('Appendix B.xls', sheet_name='07-08 FAH')
```

```python
sheets = [fah_sheet1, fah_sheet2, fah_sheet3, fah_sheet4]
years = ["94-98", "03-04", "05-06", "07-08"]
lines = ["solid", "dashed", "dotted", "dashdot"]

x_values = fah_sheet1.iloc[76:139, 0]
x_label_fruit = x_values.iloc[1:14]
dairy_first = x_values.iloc[[31]]
dairy_second = x_values.iloc[36:40]
x_label_dairy = pd.concat([dairy_first, dairy_second])
```

```python
for i, sheet in enumerate(sheets):
    men = sheet.iloc[:, 6]
    fruit_men = men.iloc[5:18]

    plt.plot(x_label_fruit, fruit_men, label=years[i], linestyle=lines[i])

plt.title("Fruit Consumption - Men")
plt.xlabel("Fruit Type")
plt.ylabel("Average Amount of Fruits (pounds)")

plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```

U.S. Food Commodity Availability by Food Source, 1994-2008 https://www.ers.usda.gov/publications/pub-details?pubid=81817

The graph above shows that for fruits overall, men's consumption was highest during 2003-2004, then 2005-2006, then 2007-2008, and finally 1994-1998. For example, berries had a huge increase in consumption from 1994-1998 to 2003-2004 peaking at around 95 pounds, and then gradually decreasing for the years after that.

The consumption for Oranges Total was around the same pounds for all years at around 90 pounds, except for 2003-2004 which was much lower than the rest at around 85 pounds

```python
for i, sheet in enumerate(sheets):
    women = sheet.iloc[:, 9]
    fruit_women = women.iloc[5:18]

    plt.plot(x_label_fruit, fruit_women, label=years[i], linestyle=lines[i])

plt.title("Fruit Consumption - Women")
plt.xlabel("Fruit Type")
plt.ylabel("Average Amount of Fruits (pounds)")

plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```

U.S. Food Commodity Availability by Food Source, 1994-2008 https://www.ers.usda.gov/publications/pub-details?pubid=81817

The graph above shows that for fruits overall, women's consumption of most fruits were relatively more similar compared to men's consumption of fruits.

Stone fruit had the lowest amounts of consumption, with rates dropping after 1994-1998 from around 85 pounds to 77.5 pounds.

The opposite can be said for grapes as in 1994-1998, around 85 pounds were consumed on average but increased drastically over the next few years.

```python
for i, sheet in enumerate(sheets):
    men = sheet.iloc[:, 6]
    men_first = men.iloc[[31]]
    men_second = men.iloc[36:40]
    dairy_men = pd.concat([men_first, men_second])

    plt.plot(x_label_dairy, dairy_men, label=years[i], linestyle=lines[i])

plt.title("Dairy Consumption - Men")
plt.xlabel("Dairy Type")
plt.ylabel("Average Amount of Dairy (pounds)")

plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```

U.S. Food Commodity Availability by Food Source, 1994-2008 https://www.ers.usda.gov/publications/pub-details?pubid=81817

There is a lot of variation for the dairy consumed by men over the years, with around highest levels of consumption from 2003-2004, then 2005-2006, 2007-2008, and finally 1994-1998. For example, the consumption of yogurt was over 86 pounds in 2003-2004, around 82 pounds for 2004-2005 and 2007-2008, and finally 78 pounds for 1994-1998.

For 2005-2006, all dairy types had around the same amount of consumption at around 82 pounds.
