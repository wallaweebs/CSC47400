```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fah_sheet1 = pd.read_excel('Appendix B.xls', sheet_name='94-98 FAH')
fah_sheet4 = pd.read_excel('Appendix B.xls', sheet_name='07-08 FAH')
```

```python
years = ["94-98", "07-08"]
lines = ["solid", "dashed", "dotted", "dashdot"]

x_values = fah_sheet1.iloc[76:139, 0]
x_label_fruit = x_values.iloc[1:14].reset_index(drop=True)
dairy_first = x_values.iloc[[31]]
dairy_second = x_values.iloc[36:40]
x_label_dairy = pd.concat([dairy_first, dairy_second])

fruit_labels = x_label_fruit.tolist()
dairy_labels = x_label_dairy.tolist()
```

```python
men_94 = fah_sheet1.iloc[:, 6]
men_07 = fah_sheet4.iloc[:, 6]
fruit_men_94 = men_94.iloc[5:18].reset_index(drop=True)
fruit_men_07 = men_07.iloc[5:18].reset_index(drop=True)

men_first_94 = men_94.iloc[[31]]
men_first_07 = men_07.iloc[[31]]

men_second_94 = men_94.iloc[36:40]
men_second_07 = men_07.iloc[36:40]
dairy_men_94 = pd.concat([men_first_94, men_second_94]).reset_index(drop=True)
dairy_men_07 = pd.concat([men_first_07, men_second_07]).reset_index(drop=True)

fruit_percents = {}
dairy_percents = {}
for i in range(len(fruit_men_94)):
        fruit_percents[fruit_labels[i]] = (fruit_men_94[i] - fruit_men_07[i]) / fruit_men_07[i] * 100
for i in range(len(dairy_men_94)):
        dairy_percents[dairy_labels[i]] = (dairy_men_94[i] - dairy_men_07[i]) / dairy_men_07[i] * 100

fruit_percents_sorted = list(sorted(fruit_percents.items(), key=lambda item: item[1]))
dairy_percents_sorted = list(sorted(dairy_percents.items(), key=lambda item: item[1]))

percents_sorted = fruit_percents_sorted + dairy_percents_sorted
sorted_labels = [item[0] for item in percents_sorted]
sorted_percentages = [item[1] for item in percents_sorted]

plt.bar(sorted_labels, sorted_percentages, color='lightgreen')

plt.title("Men's Fruit and Dairy Consumption Percents (1994-1998 to 2007-2008)")
plt.xlabel("Fruit/Dairy Type")
plt.ylabel("Percent Change")

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
women_94 = fah_sheet1.iloc[:, 9]
women_07 = fah_sheet4.iloc[:, 9]
fruit_women_94 = women_94.iloc[5:18].reset_index(drop=True)
fruit_women_07 = women_07.iloc[5:18].reset_index(drop=True)

women_first_94 = women_94.iloc[[31]]
women_first_07 = women_07.iloc[[31]]

women_second_94 = women_94.iloc[36:40]
women_second_07 = women_07.iloc[36:40]
dairy_women_94 = pd.concat([women_first_94, women_second_94]).reset_index(drop=True)
dairy_women_07 = pd.concat([women_first_07, women_second_07]).reset_index(drop=True)

fruit_percents = {}
dairy_percents = {}

for i in range(len(fruit_women_94)):
        fruit_percents[fruit_labels[i]] = (fruit_women_94[i] - fruit_women_07[i]) / fruit_women_07[i] * 100
for i in range(len(dairy_women_94)):
        dairy_percents[dairy_labels[i]] = (dairy_women_94[i] - dairy_women_07[i]) / dairy_women_07[i] * 100

fruit_percents_sorted = list(sorted(fruit_percents.items(), key=lambda item: item[1]))
dairy_percents_sorted = list(sorted(dairy_percents.items(), key=lambda item: item[1]))

percents_sorted = fruit_percents_sorted + dairy_percents_sorted
sorted_labels = [item[0] for item in percents_sorted]
sorted_percentages = [item[1] for item in percents_sorted]
plt.bar(sorted_labels, sorted_percentages, color='pink')

plt.title("Women's Fruit and Dairy Consumption Percents (1994-1998 to 2007-2008)")
plt.xlabel("Fruit/Dairy Type")
plt.ylabel("Percent Change")

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
