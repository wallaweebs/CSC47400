```python
Markdown(f"""

## 2D Scalar Field Graph

This graph shows spatial coordinates both axes (West-East and North-South positions) that represent the area around the Maunga Whau volcano in a 10m grid
The colors represent the elevation of each coordinate, with purple being lower elevations and red being higher elevations.

The volcano map is a 2D scalar field graph because it takes two spatial coordinates, x and y, and returns one scalar value (elevation) at each point.

""")
```

```python
Markdown(f"""

## 3D Scalar Field Graph

This graph shows a curved surface with varying scalar values represented by colors and a grid showing the curves of each area. There are 3 axes
representing x, y, and z coordinates. The purple regions represent lower scalar values while orange regions represent higher scalar values. 

This graph is a 3D scalar field since there are three independent spatial coordinates in the axes, the data exists in three-dimensional space,
the colors represent scalar values at each point, and the mapping of the data reveals the full 3D grid structure it has.
""")
```
